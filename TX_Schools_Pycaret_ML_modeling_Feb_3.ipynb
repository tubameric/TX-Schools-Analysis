```python
# Code snippet 1
# Importing numpy and pandas
import numpy as np
import pandas as pd
```

```python
# Code snippet 2
# Importing the dataset

url = 'https://raw.githubusercontent.com/nvamsimohan/DallasDSA/main/2021%20Texas%20High%20Schools%20Scores.csv'

data  = pd.read_csv(url)

# Displaying the data
data
```

```python
# Code snippet 3
# Displaying the data columns
data.columns
```

```python
# Code snippet 4
# Setting the dependent variable
y = 'Standard Score'
```

```python
# Code snippet 5
# Setting the noise/redundant variables & variables with more than 20% missing data
ignored_cols = ['School', 'School URL',  'District URL',  'City URL',
       'Phone', 'Home Prices Ranking', 'Community Health Ranking']
```

```python
# Code snippet 6
# Setting the categorical variables
cat_cols = ['District', 'City',
       'Zip', 'County', 'Is Title I', 'Is Charter', 'Is Magnet',
       'Is Virtual']
```

```python
# Code snippet 7
# Setting the numerical variables
num_cols = ['Number Students', 'Number Full-time Teachers',
       'Student/Teacher Ratio', 'Percent Free/Disc Lunch',
       'Percent Homes Rented', 'Percent of Unemployment',
       'Percent Parents with Masters and above education',
       'Percent parents with less than High School education',
       'Percent Home Incomes with less than Median income']
```

```python
# Code snippet 8
# Importing the regression libraries from pycaret
from pycaret.regression import *
```

```python
# Code snippet 9
# Setting/configuring the pycaret ML experiment
regression_setup = setup(data,
                         target = y,
                         categorical_features = cat_cols,
                         numeric_features = num_cols,
                         ignore_features = ignored_cols)
```
