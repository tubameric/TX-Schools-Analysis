```python
# Code snippet 0
url='https://raw.githubusercontent.com/nvamsimohan/DallasDSA/main/2021%20Texas%20High%20Schools%20Scores.csv'

import pandas as pd
data = pd.read_csv(url)
data.columns
```

```python
# Code snippet 1
predictors = data[['District', 'City',
       'Zip', 'County',  'Is Title I', 'Is Charter', 'Is Magnet',
       'Is Virtual',
       'Student/Teacher Ratio', 'Percent Free/Disc Lunch',
       'Percent Homes Rented', 'Percent of Unemployment',
       'Percent Parents with Masters and above education',
       'Percent parents with less than High School education',
       'Percent Home Incomes with less than Median income',
       'Home Prices Ranking', 'Community Health Ranking', 'Standard Score']]

predictors.shape
```

```python
# Code snippet 3
import sketch
```

```python
predictors.sketch.howto("get descriptive statistics of this dataframe")
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Get summary statistics of the dataframe
predictors.describe()

# Get descriptive data of the dataframe
predictors.info()

# Get unique values for each column
for col in predictors.columns:
    print(col, predictors[col].unique())

# Get head sample of the dataframe
predictors.head()

# Get quantiles for each column
for col in predictors.columns:
    print(col, np.quantile(predictors[col], [0, 0.25, 0.5, 0.75, 1]))
```

```python
predictors.sketch.howto("derive age in years from birthdate field")
```

```python
# Convert birthdate column to datetime format
predictors['Birthdate'] = pd.to_datetime(predictors['Birthdate'])

# Calculate current date
current_date = pd.to_datetime('today')

# Create new column for age in years
predictors['Age'] = (current_date - predictors['Birthdate']).astype('<m8[Y]')

# Print head of dataframe to check results
print(predictors.head())
```

```python
predictors.sketch.howto("get correlation values between 'Home Prices Ranking','Community Health Ranking','Standard Score'")
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Create a new dataframe with only the desired columns
corr_df = predictors[['Home Prices Ranking', 'Community Health Ranking', 'Standard Score']]

# Calculate the correlation matrix
corr_matrix = corr_df.corr()

# Print the correlation values
print(corr_matrix)
```

```python
predictors.sketch.howto("get missing data proportion for every variable")
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Calculate the proportion of missing data for each variable
missing_prop = predictors.isnull().sum() / len(predictors)

# Print the results
print("Proportion of missing data for each variable:")
print(missing_prop)
```

```python
predictors.sketch.howto("change zip to categorical")
```

```python
# Convert Zip column to categorical data type
predictors['Zip'] = predictors['Zip'].astype('category')

# Verify change
print(predictors.dtypes)
```

```python
predictors.sketch.howto("plot a chart between standard scores and charter and non-charter schools")
```

```python
import matplotlib.pyplot as plt

# Create a new dataframe with only the columns 'Is Charter' and 'Standard Score'
charter_df = predictors[['Is Charter', 'Standard Score']]

# Group the dataframe by 'Is Charter' and calculate the mean of 'Standard Score'
charter_mean = charter_df.groupby('Is Charter')['Standard Score'].mean()

# Plot a bar chart with the mean standard scores for charter and non-charter schools
plt.bar(charter_mean.index, charter_mean.values)
plt.xlabel('School Type')
plt.ylabel('Mean Standard Score')
plt.title('Mean Standard Scores for Charter and Non-Charter Schools')
plt.show()
```
